```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
train_history = pd.read_excel("result/train_history.xlsx")
train_history.head()
```

```python
plt.figure()
epoch = train_history["epoch"]
train_corr = train_history["train_corr"]
val_corr = train_history["val_corr"]

plt.plot(epoch, train_corr, c='red', label='train')
plt.plot(epoch, val_corr, c='blue', label='test')

plt.legend(loc='best')
plt.xlabel("epoch")
plt.ylabel("pearson correlation")

plt.show()
```
